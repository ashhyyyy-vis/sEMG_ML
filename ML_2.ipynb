```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC
```

```python
df=pd.read_csv('/content/emg_features_rms_mav_wl.csv')
pf=pd.read_csv('/content/emg_features_rms_mav_wl_var.csv')
```

```python
df.shape
```

```python
feature_cols = [col for col in df.columns if col.startswith("ch")]
X = df[feature_cols]
x=pf[feature_cols]
X.shape
```

```python
y = df["gesture"]
Y=pf["gesture"]
y.shape
```

```python
subjects = df["subject"]
SUBJECTS = pf["subject"]
subjects.shape
```

```python
scaler = StandardScaler()
```

```python

train_subjects = list(range(1, 21))
test_subjects  = list(range(21, 26))
train_mask = subjects.isin(train_subjects)
test_mask  = subjects.isin(test_subjects)
X_train = X[train_mask]
X_test  = X[test_mask]

x_train = x[train_mask]
x_test  = x[test_mask]

y_train = y[train_mask]
y_test  = y[test_mask]

Y_train = Y[train_mask]
Y_test  = Y[test_mask]

```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled  = scaler.transform(x_test)
```

```python
knn = KNeighborsClassifier(
    n_neighbors=5,
    metric="euclidean"
)
```

```python
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)
acc = accuracy_score(y_test, y_pred)
print("Test accuracy:", acc)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
knn.fit(x_train_scaled, Y_train)
Y_pred = knn.predict(x_test_scaled)
acc = accuracy_score(Y_test, Y_pred)
print("Test accuracy:", acc)
cm = confusion_matrix(Y_test, Y_pred)
print(cm)
```

```python
svm = SVC(
    kernel="rbf",
    C=10,
    gamma=0.1
)
```

```python
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
```

```python
svm_acc = accuracy_score(y_test, y_pred_svm)
print("SVM Test accuracy:", svm_acc)
```

```python
cm_svm = confusion_matrix(y_test, y_pred_svm)
print(cm_svm)
```

```python
svm.fit(x_train_scaled, Y_train)
Y_pred_svm = svm.predict(x_test_scaled)
```

```python
svm_acc = accuracy_score(Y_test, Y_pred_svm)
print("SVM Test accuracy:", svm_acc)
```

```python
cm_svm = confusion_matrix(y_test, y_pred_svm)
print(cm_svm)
```
